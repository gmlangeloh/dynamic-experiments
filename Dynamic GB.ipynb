```python
#How to use the current dynamic_gb implementation

load("benchmarks.sage")
load("dynamic-perry.spyx")

#Loading some common instances (cyclic family)
#Benchmark(I), for I some ideal previously loaded in Sage also works
c4 = Benchmark("./instances/cyclicn4.ideal") #The sensitivity analysis algorithm has problems with this instance!
c4h = Benchmark("./instances/cyclicnh4.ideal")
c5 = Benchmark("./instances/cyclicn5.ideal")
c5h = Benchmark("./instances/cyclicnh5.ideal")
c6 = Benchmark("./instances/cyclicn6.ideal")
c6h = Benchmark("./instances/cyclicnh6.ideal")
c7 = Benchmark("./instances/cyclicn7.ideal")
c7h = Benchmark("./instances/cyclicnh7.ideal")

#Using Perry's restricted dynamic algorithm from Caboara and Perry (2014)
_=dynamic_gb(c4.ideal.gens(), strategy="sugar")

#Using sensitivity analysis based algorithm
_=dynamic_gb(c4h.ideal.gens(), strategy="sugar", simplex=True)

#Optionally, we can use max_calls to switch to Perry's algorithm after a set number of iterations
_=dynamic_gb(c5.ideal.gens(), strategy="sugar", simplex=True, max_calls=10)

#Options static (classic Buchberger algorithm), unrestricted (Gritzmann-Sturmfels, 1993), 
#random and perturbation are also available - just set the corresponding parameters to true instead of simplex
```
